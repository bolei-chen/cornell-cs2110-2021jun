```python
import matplotlib.pyplot as plt
import random as rd
import math
```

```python
reduced_fraction_count = 0.0
total_count = 0.0
ratio = 0.0
ratio_list = []
iteration_list = []
standard_list = []
for i in range(10000):
    num1 = rd.randint(-99999, 99999)
    num2 = rd.randint(-99999, 99999)
    gcd = math.gcd(num1, num2)
    if gcd == 1:
        reduced_fraction_count += 1
    total_count += 1
    ratio = reduced_fraction_count / total_count
    ratio_list.append(ratio)
    iteration_list.append(i)
    standard_list.append(6 / math.pow(math.pi, 2))
```

```python
plt.figure(figsize=(15, 9))
plt.plot(iteration_list, ratio_list)
plt.plot(iteration_list, standard_list)
plt.ylabel("ratio")
plt.xlabel("iterations")
plt.show()
```
